```python
import os
import shutil


def copy_files(source_file_path, destination_directory):
    # Ensure source and destination directories exist
    if not os.path.exists(source_file_path):
        print("Source or destination directory does not exist.")
        return

    directory_path = os.path.dirname(destination_directory)
    if not os.path.exists(directory_path):
        os.makedirs(directory_path)

    shutil.copy(source_file_path, destination_directory)
    print(f"Copied {source_file_path} to {destination_directory}")


for class_imgs_path in [
    bloco_alv_test_set,
    bloco_chap_test_set,
    reboco_test_set,
    tijolo_alv_test_set,
    tijolo_chap_test_set,
]:
    for img_path in class_imgs_path:
        copy_files(img_path, img_path.replace("all_data", "test_set"))


for class_imgs_path in [
    bloco_alv_train_set,
    bloco_chap_train_set,
    reboco_train_set,
    tijolo_alv_train_set,
    tijolo_chap_train_set,
]:
    for img_path in class_imgs_path:
        copy_files(img_path, img_path.replace("all_data", "train_set"))
```
